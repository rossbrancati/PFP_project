**Leave one out cross validation with Naive Bayes**

Tested on individual strides and means of strides

```python
#import packages
import numpy as np
import pandas as pd

#import Naive Bayes classifier
from sklearn.naive_bayes import MultinomialNB

#import mean squared error metric
from sklearn.metrics import mean_squared_error

#import warnings and ignore them
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

```python
#import data
data_strides = pd.read_csv('/Users/rossbrancati/Desktop/PFP_Project/machine_learning/machine_learning_data/PC_scores_matrix/strides/first_three_pcs_strides.csv')
data_strides = data_strides.drop(columns=['Unnamed: 0'])
data_means = pd.read_csv('/Users/rossbrancati/Desktop/PFP_Project/machine_learning/machine_learning_data/PC_scores_matrix/means/first_three_pcs_means.csv')
data_means = data_means.drop(columns=['Unnamed: 0'])
```

```python
#add the minimum value to the data_strides and data_means dataframes so that the naive bayes algorithm won't throw and error
data_strides += (data_strides.values.min()*-1)
data_means += (data_means.values.min()*-1)
```

```python
#import labels 
labels_strides = pd.read_csv('/Users/rossbrancati/Desktop/PFP_Project/machine_learning/machine_learning_data/PC_scores_matrix/strides/labels.txt', sep='\t')
labels_means = pd.read_csv('/Users/rossbrancati/Desktop/PFP_Project/machine_learning/machine_learning_data/PC_scores_matrix/means/labels.txt', sep='\t')
```

**Define a function with the following arguments:**
* data: the dataframe<br>
* labels: the labels of the dataframe<br>
* ids: the id of the data point being left out<br>
* dropped_cols: any columns to exclude from analysis<br>
* alpha: smoothing parameter<br>
* fit_prior: True = learn class probabilities, False = don't learn class probabilities

```python
def LOOCV_prediction(data, labels, ids, alpha, fit_prior):
    
    #step 1: generate the model
    clf = MultinomialNB(alpha=alpha, fit_prior=fit_prior)
    
    #step 2: assign the training and testing data
    X_train = data.drop([ids])
    X_test = data.iloc[[ids]]
    
    #step 3: assign the training and testing labels
    y_train = labels.drop([ids])
    y_test = labels.iloc[ids]
    
    #step 4: fit the model with training data
    clf.fit(X_train, y_train)
    
    #step 5: generate prediction of the testing dataset
    prediction = clf.predict(X_test)
    
    return prediction
```

**Naive Bayes for Mean Data**

```python
#now we will try looping over some different hyperparameter values and storing MSE to see which 
#convert the labels data frame to an array
y_true = labels_means.to_numpy()

#hyperparameters generate the best predictions
alpha = [0.001, 0.01, 0.1, 1, 10]
fit_prior = [False, True]

#create an array to store results
results = np.zeros((len(alpha)*len(fit_prior), 3))

#iteration variable to determine which iteration you are on
iteration = 0

#loop over each value of the number of estimators
for a in alpha:
    #loop over each value of the max depth
    for f in fit_prior:
        #generate an empty predictions array
        y_pred = np.array([])
        #loop over each row of the data frame
        for ids in range(len(data_means)):
            #call function to train model and generate a prediction
            current_prediction = LOOCV_prediction(data_means, labels_means, ids, alpha=a, fit_prior=f)
            #append the prediction to the predictions array
            y_pred = np.append(y_pred, current_prediction)
        #calculate mean squared error of the predictions vs. ground truth
        mse = mean_squared_error(y_true, y_pred)
        
        #add the hyperparameters and MSE to the results table
        results[iteration, 0] = a
        results[iteration, 1] = fit_prior.index(f)
        results[iteration, 2] = mse
        
        #add an iteration to the overall loop
        iteration = iteration+1
        
#convert the results array to a pandas dataframe
results_means = pd.DataFrame(results, columns=['Alpha', 'Fit Prior', 'Mean Squared Error'])

#display the results
results_means
```
