Tested on individual strides and means of strides

```python
#import packages
import numpy as np
import pandas as pd

#import random forest classifier
from sklearn.ensemble import RandomForestClassifier

#import mean squared error metric
from sklearn.metrics import mean_squared_error

#import warnings and ignore them
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

**Define a function with the following arguments:**
* data: the dataframe<br>
* labels: the labels of the dataframe<br>
* ids: the id of the data point being left out<br>
* dropped_cols: any columns to exclude from analysis<br>
* num_estimators: hyperparameter - number of trees in the forest<br>
* max_depth: maximum depth of the tree

```python
def LOOCV_prediction(data, labels, ids, num_estimators, max_depth):
    
    #step 1: generate the model
    clf = RandomForestClassifier(n_estimators=num_estimators, max_depth=max_depth)
    
    #step 2: assign the training and testing data
    X_train = data.drop([ids])
    X_test = data.iloc[[ids]]
    
    #step 3: assign the training and testing labels
    y_train = labels.drop([ids])
    y_test = labels.iloc[ids]
    
    #step 4: fit the model with training data
    clf.fit(X_train, y_train)
    
    #step 5: generate prediction of the testing dataset
    prediction = clf.predict(X_test)
    
    return prediction
```

This cell is a test of the function above

```python
#first try just looping over all ids with the same hyperparameters and generating an array of predictions
#generate an empty predictions array
y_pred = np.array([])
#loop over each row of the data frame
for ids in range(len(data_means)):
    #call function to train model and generate a prediction
    current_prediction = LOOCV_prediction(data_means, labels_means, ids, num_estimators=100, max_depth=3)
    #append the prediction to the predictions array
    y_pred = np.append(y_pred, current_prediction)

#calculate mean squared error of the predictions vs. ground truth
y_true = labels_means.to_numpy()
mean_squared_error(y_true, y_pred)
    
```

**Random Forest Classifier for Mean Data**

```python
#now we will try looping over some different hyperparameter values and storing MSE to see which 
#convert the labels data frame to an array
y_true = labels_means.to_numpy()

#hyperparameters generate the best predictions
estimators = [5, 10, 50, 100, 200]
depths = [3, 5, 10, 20]

#create an array to store results
results = np.zeros((len(estimators)*len(depths), 3))

#iteration variable to determine which iteration you are on
iteration = 0

#loop over each value of the number of estimators
for e in estimators:
    #loop over each value of the max depth
    for d in depths:
        #generate an empty predictions array
        y_pred = np.array([])
        #loop over each row of the data frame
        for ids in range(len(data_means)):
            #call function to train model and generate a prediction
            current_prediction = LOOCV_prediction(data_means, labels_means, ids, num_estimators=e, max_depth=d)
            #append the prediction to the predictions array
            y_pred = np.append(y_pred, current_prediction)
        #calculate mean squared error of the predictions vs. ground truth
        mse = mean_squared_error(y_true, y_pred)
        
        #add the hyperparameters and MSE to the results table
        results[iteration, 0] = e
        results[iteration, 1] = d
        results[iteration, 2] = mse
        
        #add an iteration to the overall loop
        iteration = iteration+1
        
#convert the results array to a pandas dataframe
results_means = pd.DataFrame(results, columns=['Number of Estimators', 'Max Depth', 'Mean Squared Error'])

#display the results
results_means
```
